```python
@np.vectorize 
def remove_dollar(label:str):
    return float(label.replace("$","").replace(",",""))
    
listings_new=listings.assign(price=remove_dollar(listings.price))
listings_new.price
```

```python
listings_new\
.groupby("room_type",dropna=False)\
.id.count()\
.reset_index()\
.rename(columns={"id":"count"})
```

```python
Variation=listings_new\
.groupby("room_type",dropna=False)\
.id.count()\
.reset_index()\
.rename(columns={"id":"count"})

Variation["previous"]= pd.Series(["17248","1276","8923","200"]).astype("int64")
Variation["variation_percent"]=100*(((Variation["count"]-Variation["previous"])/Variation["previous"]).round(2))
Variation
```

```python
Percent_variation=((Variation["count"]-Variation["previous"])/Variation["count"]).round(2)
(listings_new.room_type.value_counts()/len(listings_new.room_type)).round(2)
```

```python
mean_price=listings_new.agg({"price":[np.mean]})
mean_price
```

```python
old_price=105.107643
print(f"The mean variation on the price is: {(100*(mean_price-old_price)/old_price).round(2)}" + "%")
```

```python
listings_new.groupby('room_type').agg({'price': [np.mean]})
```

```python
price2021=pd.Series({"Entire home/apt":114.339170,"Hotel room":96.080721,"Private room":90.041690,"Shared room":38.740000})
Mean_type=listings_new.groupby('room_type').agg({'price': [np.mean]})
Mean_type["mean_2021"]=price2021
Mean_type.columns=["mean2022","mean2021"]
Mean_type["difference"]=Mean_type["mean2022"]-Mean_type["mean2021"]
Mean_type["difference_perc"]=100*(Mean_type["difference"]/Mean_type["mean2021"]).round(2)
Mean_type
```

```python
listings_new2 = listings_new\
  .groupby('room_type')\
  .agg({
      'price': ['count', 'mean']
  })\
  .reset_index()

listings_new2['count_perc'] = listings_new2['price']['count'] / listings_new2['price']['count'].sum() * 100

listings_new2
```

```python
pivot_df = listings_new\
    .groupby(['property_type','room_type'])\
    .price\
    .mean()\
    .round(0)\
    .reset_index()

pivot_df
```

```python
pivot_table = pivot_df\
    .pivot(
        columns='room_type',
        values='price',
        index=['property_type']
    )\
    .reset_index()

pivot_table
```
