```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Historical%20gold%20price%20from%201791%20to%202020%20in%20USD/gold_price_1791_2020_yearly_denvergold.csv')
```

```python
f,ax= plt.subplots(figsize=(5,5))
sns.heatmap(data.corr(), annot= True,  fmt='.1f' , ax=ax)
plt.show()
```

```python
f,ax= plt.subplots(figsize=(5,5))
sns.heatmap(data.corr(), annot= True,  fmt='.1f' , ax=ax)
plt.show()
```

```python
data = data.rename(columns={'Nominal ounce closing price USD': 'Nominal'})
data = data.rename(columns={'Real ounce closing price USD': 'Real'})
```

```python
data.columns
```

```python
data.Nominal.plot(kind = 'line', color = 'g',label = 'Nominal ounce closing price USD',linewidth=3,alpha = 1,grid = True,linestyle = '-')

plt.legend(loc='upper right')   
plt.xlabel('time')              
plt.ylabel('Nominal ounce closing price USD')
plt.title('The time dependent change ')          
plt.show()
```
