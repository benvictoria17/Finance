```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Major%20social%20media%20stock%20prices%202012-2022/social%20media%20stocks%202012-2022.csv")
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
df['month_year'] = df['Date'].dt.to_period('M')
symbols = ['FB','ETSY','TWTR','SNAP','PINS']

for symbol in symbols:
    df[df['Symbol']==symbol].groupby('month_year').mean()['Adj Close'].plot(
        figsize = (12,8), grid = True, label = symbol, fontsize = 15)
    plt.legend()
    
plt.title('Stocks prices changes', fontsize = 15)
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Price in US', fontsize = 15)
plt.show()
```

```python
symbols = ['FB','ETSY','TWTR','SNAP','PINS']

for symbol in symbols:
    df[(df['Symbol']==symbol)&(df['year']>=2020)].groupby('month_year').mean()['Adj Close'].plot(
        figsize = (12,8), grid = True, label = symbol, fontsize = 15)
    plt.legend()
plt.title('Stocks prices changes', fontsize = 15)
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Price in US', fontsize = 15)
plt.show()
```

```python
symbols = ['FB','ETSY']

for symbol in symbols:
    df[(df['Symbol']==symbol)&(df['month_year']>='2021-10')].groupby('Date').mean()['Adj Close'].plot(
        figsize = (12,8), grid = True, label = symbol, fontsize = 15)
    plt.legend()
plt.title('Stocks prices changes', fontsize = 15)
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Price in US', fontsize = 15)
plt.show()
```

```python
symbols = ['FB']

for symbol in symbols:
    volume = df[(df['Symbol']==symbol)&(df['Date'].between('2022-02-01','2022-02-15'))].groupby('Date').mean()['Volume'].plot(
        figsize = (12,8), grid = True, label = symbol, fontsize = 15)
    plt.legend()
plt.title('Volume changes', fontsize = 15)
plt.xlabel('Date', fontsize = 15)
plt.ylabel('Volume', fontsize = 15)
plt.show()
```
