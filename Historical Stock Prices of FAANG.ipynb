```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
amazon = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Amazon.csv")
apple = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Apple.csv")
facebook = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Facebook.csv")
google = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Google.csv")
microsoft = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Microsoft.csv")
netflix = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Netflix.csv')
tesla = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Tesla.csv')
uber = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Uber.csv')
walmart = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Walmart.csv')
zoom = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Historical%20Stock%20Price%20of%20(FAANG%20%2B%205)%20companies/Zoom.csv')
amazon['company'] = 'amazon'
apple['company'] = 'apple'
facebook['company'] = 'facebook'
google['company'] = 'google'
microsoft['company'] = 'microsoft'
netflix['company'] = 'netflix'
tesla['company'] = 'tesla'
uber['company'] = 'uber'
walmart['company'] = 'walmart'
zoom['company'] = 'zoom'
```
