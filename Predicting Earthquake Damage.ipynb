```python
surface_ordinal=earthquake_data.groupby(['land_surface_condition'])['age'].mean().to_dict()
surface_ordinal
```

```python
earthquake_data['surface_condition']=earthquake_data['land_surface_condition'].map(surface_ordinal)
earthquake_data.head()
```

```python
final_data=earthquake_data.drop(labels=['land_surface_condition', 'foundation_type', 'roof_type',
       'ground_floor_type', 'other_floor_type', 'position',
       'plan_configuration','legal_ownership_status'],axis=1)
final_data.shape
```

```python
X=final_data.drop('damage_grade',axis=1)
y=final_data['damage_grade']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
final_data.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier().fit(X_train,y_train)
y_predicted=classifier.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(y_test,y_predicted)
```

```python
metrics.confusion_matrix(y_test,y_predicted)
```

```python
def correlation(dataset, threshold):
    col_corr = set() 
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]  
                col_corr.add(colname)
    return col_corr
correlation(X_train,0.70)
```
