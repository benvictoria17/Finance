```python
import pandas as pd
cost_of_living = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Cost%20of%20Living/cost-of-living.csv")
```

```python
cost_of_living = cost_of_living.T
cost_of_living.head()
```

```python
import seaborn as sns

sns.boxplot(y=cost_of_living[0][1:].astype(float))
```

```python
print(cost_of_living[0].keys()[1:])
```

```python
import matplotlib.pyplot as plt
import numpy as np 
height = cost_of_living[0][1:].astype(float)
bars = cost_of_living[0].keys()[1:]
y_pos = np.arange(len(bars))

plt.figure(figsize=(20,10))
 
plt.bar(y_pos, height)
 
plt.xticks(y_pos, bars, rotation=90)
 
plt.show()
```
