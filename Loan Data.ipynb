```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Loan%20Data/loan_data.csv")
data.head()
```

```python
data.describe()
```

```python
data.corr()
```

```python
sns.histplot(x='fico',data= data, kde=True,bins = 20);
```

```python
plt.figure(figsize=(11,5))
sns.countplot(x='purpose', hue='not.fully.paid', data=data).set(title="Purpose for taking loan");
```

```python
sns.jointplot(x='fico', y='int.rate', data=data, marker='x');
```

```python
data = pd.get_dummies(data, columns=['purpose'], drop_first=True)
from sklearn.model_selection import train_test_split
x = data.drop('not.fully.paid',axis=1)
y = data['not.fully.paid']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=3)
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
predictions = dtree.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
accuracy_score(y_test, predictions)
```
