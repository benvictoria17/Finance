```python
def nullremoval(dataframe_app,percent=0.30):     #function to accept the dataframe and percent value.
    df_app = dataframe_app.copy()                #creating a copy of the dataframe
    colnames = (df_app.isnull().sum()/len(df_app))  # finding out empty or blank columns
    colnames = list(colnames[colnames.values>=percent].index)  # checking if the values are greater then or equal to the percent value and then storing it in columns.
    df_app.drop(labels = colnames,axis =1,inplace=True)  
    print("Total number of columns dropped: ",len(colnames))    
    return df_app
app_d = nullremoval(app_d,percent = 0.30)
```
