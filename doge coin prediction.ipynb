```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/benvictoria17/Finance/master/dataset/Dogecoin%20INR%20Dataset%202017-2022/DOGE-INR.csv')
data.head()
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data.info()
```

```python
data.groupby(['High' , 'Low']).mean()
```

```python
data.isnull().sum()
```

```python
data['Open'] = data['Open'].fillna(data['Open'].mean())
data['High'] = data['High'].fillna(data['High'].mean())
data['Low'] = data['Low'].fillna(data['Low'].mean())
data['Close'] = data['Close'].fillna(data['Close'].mean())
data['Adj Close'] = data['Adj Close'].fillna(data['Adj Close'].mean())
data['Volume'] = data['Volume'].fillna(data['Volume'].mean())
data.dtypes
```
