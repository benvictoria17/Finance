```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df1 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/master/dataset/NSE%20India%20stocks%20(Indices)/banknifty.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/master/dataset/NSE%20India%20stocks%20(Indices)/nifty50.csv")
```

```
df1['date']=pd.to_datetime(df1['date'].astype(str), format='%Y%m%d')
df2['date']=pd.to_datetime(df2['date'].astype(str), format='%Y%m%d')
```

```
date_df1=df1.groupby('date').mean()
date_df2=df2.groupby('date').mean()
```

```
date_df1.head()
```

```
date_df1.describe()
```

```
date_df2.head()
```

```
plt.figure(figsize=(20,7))
plt.plot(date_df1,color='blue')
plt.title('Bank Nifty Graph')
```

```
plt.figure(figsize=(20,7))
plt.plot(date_df2,color='blue')
plt.title('Nifty50 Graph')
```
