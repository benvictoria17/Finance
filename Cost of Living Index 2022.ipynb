```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/main/dataset/Cost%20of%20Living%20Index%202022/Cost_of_Living_Index_2022.csv")
```

```python
dff = list(df.Country[:])
df.head()
```

```python
top_10_resturant = df.sort_values(by = ['Restaurant Price Index'],ascending=False).head(10)
top_10_resturant
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=top_10_resturant['Country'], y=top_10_resturant['Restaurant Price Index'])
plt.xticks(rotation= 45)
plt.xlabel('Country')
plt.ylabel('Restaurant Price Index')
plt.title('top_10_resturant')
```

```python
top_10_grocery = df.sort_values(by = ['Groceries Index'],ascending=False).head(10)
top_10_grocery
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=top_10_grocery['Country'], y=top_10_grocery['Groceries Index'])
plt.xticks(rotation= 45)
plt.xlabel('Country')
plt.ylabel('Groceries Index')
plt.title('top_10_grocery')
```

```python
last_10_grocery = df.sort_values(by = ['Groceries Index'],ascending=False).tail(10)
last_10_grocery
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=last_10_grocery['Country'], y=last_10_grocery['Groceries Index'])
plt.xticks(rotation= 45)
plt.xlabel('Country')
plt.ylabel('Groceries Index')
plt.title('last_10_grocery')
```

```python
top_10_rent = df.sort_values(by = ['Rent Index'],ascending=False).head(10)
top_10_rent
```

```python
top_10_rent = df.sort_values(by = ['Rent Index'],ascending=False).head(10)
top_10_rent
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=top_10_rent['Country'], y=top_10_rent['Rent Index'])
plt.xticks(rotation= 45)
plt.xlabel('Country')
plt.ylabel('Rent Index')
plt.title('top_10_rent')
```

```python
last_10_rent = df.sort_values(by = ['Rent Index'],ascending=False).tail(10)
last_10_rent
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=last_10_rent['Country'], y=last_10_rent['Rent Index'])
plt.xticks(rotation= 45)
plt.xlabel('Country')
plt.ylabel('Rent Index')
plt.title('last_10_rent')
```

```python
rent_grocery_resturant_top = pd.concat([top_10_rent,top_10_grocery,top_10_resturant])
rent_grocery_resturant_top.value_counts()
```
