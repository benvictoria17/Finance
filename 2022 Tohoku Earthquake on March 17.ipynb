```python
import os
import numpy as np
import pandas as pd
import random
import folium
from folium import plugins
import seaborn as sns
import cv2
import matplotlib.pyplot as plt
import IPython
import time
import plotly.express as px
```

```python
data0=pd.read_csv('https://raw.githubusercontent.com/benvictoria17/DataAnalytics/master/dataset/Earthquake%20in%20Japan/japan_20220319_query.csv')
data0=data0.sort_values('time',ascending=True).reset_index(drop=True)
data0[-5:]
```

```python
print(data0.columns.tolist())
```

```python
data0=data0.fillna('N')
data1=data0[['time','latitude','longitude','depth','mag','type']]
data1[-5:]
```

```python
from datetime import datetime as dt
data2a=data1.copy()
date=[]
for item in data1['time']:
    date+=[dt.strptime(item[0:-5],'%Y-%m-%dT%H:%M:%S')]
data2a['date']=date
data2a
```

```python
import plotly.express as px
fig=px.scatter(x=data2a['date'],y=data2a['mag'])
fig=px.scatter(data2a, x="date", y="mag")
fig.update_layout(title='Magnituide of earthquake in Tohoku')
fig.show()
```

```python
data2b=data2a[data2a['mag']>4.9].reset_index(drop=True)
display(data2b)
```

```python
data2c=data2b[['latitude','longitude']]
display(data2c)
```

```python
for i in range(len(data2b)):
    item=str(data2b.loc[i:i,'date'].tolist()[0])
    eq_map = folium.Map(location=[37.6,141.6],tiles='Stamen Terrain',zoom_start=6.0,min_zoom=2.0)
    eq_map.add_child(plugins.HeatMap(data2c.loc[i:i]))
    eq_map.save("./"+str(item)+".html")
```
