```
import numpy as np 
import pandas as pd 
df = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/Finance/master/dataset/NFT%20History%20Sales/NFT_Sales.csv")
df.head()
```

```
df.describe()
```

```
import matplotlib.pyplot as plt

def plot_df(df, x, y, title="", xlabel='Date', ylabel='NFT', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df, x=df.index, y=df.Number_of_Sales, title='Daily NFT Sales.')    
```

```
plot_df(df, x=df.index, y=df.AverageUSD_cum, title='Average Sale Price.')    
```

```
plot_df(df, x=df.index, y=df.Active_Market_Wallets_cumsum, title='Active Market Wallets')    
```
